```python
from langgraph.graph import StateGraph
from typing import TypedDict, Literal
from IPython.display import display, Image

class State(TypedDict):
    val: str

def search_wall_node(state): ...
def backward_some_meters_node(state): ...
def forward_node(state): ...
def rotate_90_node(state): ...

def retry_edge(state) -> Literal["search_wall", "forward"]: ...

graph_builder = StateGraph(State)
graph_builder.add_node("search_wall", search_wall_node)
graph_builder.add_node("backward_some_meters", backward_some_meters_node)
graph_builder.add_node("forward", forward_node)
graph_builder.add_node("rotate", rotate_90_node)

graph_builder.add_edge("__start__", "search_wall")
graph_builder.add_edge("search_wall", "backward_some_meters")
graph_builder.add_edge("backward_some_meters", "forward")
graph_builder.add_edge("forward", "rotate")
graph_builder.add_conditional_edges("rotate", retry_edge)

graph = graph_builder.compile()
image = graph.get_graph().draw_mermaid_png()
```

```python
display(Image(image))
with open("../images/practice_1_rombus.png", mode="wb") as f:
    f.write(image)
```
